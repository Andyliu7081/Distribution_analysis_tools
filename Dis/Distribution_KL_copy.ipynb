```python
import numpy as np
import matplotlib.pyplot as plt

def calc_kl_divergence(mu_p, sigma_p, mu_q, sigma_q):
    # Generate Gaussian data (for visualization purposes)
    s_p = np.random.normal(mu_p, sigma_p, 100000)
    s_q = np.random.normal(mu_q, sigma_q, 100000)

    # Plot
    plt.hist(s_p, bins=30, density=True, alpha=0.5, label='p')
    plt.hist(s_q, bins=30, density=True, alpha=0.5, label='q')
    plt.legend(loc='upper right')
    plt.show()

    # Compute KL divergence
    D_KL = np.log(sigma_q / sigma_p) + (sigma_p**2 + (mu_p - mu_q)**2) / (2 * sigma_q**2) - 0.5
    return D_KL

def calc_left_kl_divergence(mu_p, sigma_p):
    # mu_q, sigma_q = 2.0, 4.5
    # mu_q, sigma_q = 0.361, 1.465
    mu_q, sigma_q = 0.3158, 1.6632 # grid search result

    return calc_kl_divergence(mu_p, sigma_p, mu_q, sigma_q)


def calc_right_kl_divergence(mu_p, sigma_p):
    # mu_q, sigma_q = 3.652, 6.168
    mu_q, sigma_q = 4.7368, 7.3947 # grid search result
    return calc_kl_divergence(mu_p, sigma_p, mu_q, sigma_q)
```

```python
left_mean = 0.3158
left_std = 1.6632

right_mean = 4.7368
right_std = 7.3947

left = calc_left_kl_divergence(left_mean, left_std)
right = calc_right_kl_divergence(right_mean, right_std)
value = (left + right) / 2
print(value)
```
